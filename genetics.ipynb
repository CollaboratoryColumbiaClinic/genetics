```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
import scipy 

sns.set(font="monospace")
% matplotlib inline
data=pd.read_csv("index1-genes.csv",index_col=0)
data_array=np.asarray(data)
```

```python
distance=scipy.spatial.distance.pdist(data_array,metric ="jaccard")

output=scipy.cluster.hierarchy.linkage(distance,method="ward")
```

```python
pd.DataFrame(output).to_csv("output1.csv")
```

```python
dict=scipy.cluster.hierarchy.dendrogram(output)
```

```python
label=list(zip(dict["leaves"],dict["color_list"]))
pd.DataFrame(label).to_csv("output2.csv",header=["Gene","Label"],index=None)
```

```python
label_df=pd.DataFrame(label)
label_df
```

```python
label_df_g=label_df[label_df[1]=="g"]
label_df_g
```

```python
index_g=np.asarray(label_df_g[0])
index_g
```

```python
data_array_g=data_array[index_g,:]
data_array_g.shape
```

```python
distance_g=scipy.spatial.distance.pdist(data_array_g,metric ="jaccard")

output_g=scipy.cluster.hierarchy.linkage(distance_g,method="ward")
```

```python
dict_g=scipy.cluster.hierarchy.dendrogram(output_g,color_threshold=1.5)
```
