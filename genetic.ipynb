```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
% matplotlib inline
```

```python
data=pd.read_csv("edgelist.csv")
data.head(
```

```python
similarity=data["Weight"]
plt.figure()
similarity.hist(bins=50)
```

```python
minmax=MinMaxScaler()
similarity_scale=minmax.fit_transform(similarity)
plt.hist(similarity_scale,bins=50)
data.loc[:,'Weight']=similarity_scale
data.head()
```

```python
data.to_csv("scale.csv",index=False)
data
```
